```python
with open('D10.txt') as f:
    raw_lines = f.readlines()

lines = []
for line in raw_lines:
    lines.append(line.strip())
```

```python
len(lines)
```

```python
x = 1
cycle = 1
signal_strength = 0

for i in range(len(lines)):
    
    instruct = lines[i]
    
    if instruct == 'noop':
        cycle += 1
        
    elif instruct[0:4] == 'addx':
        x += int(instruct[5::])
        cycle += 2
    
    signal_strength = x * cycle
    
    print('x:', x, ', cycle:', cycle, ', signal_strength:', signal_strength)
        
```
