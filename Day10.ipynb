```python
import pandas as pd
```

```python
with open('D10.txt') as f:
    raw_lines = f.readlines()

lines = []
for line in raw_lines:
    lines.append(line.strip())
```

```python
expanded_lines = []
for line in lines:
    if line[0:4] == 'addx':
        expanded_lines.append('addx 0')
        expanded_lines.append(line)
    else:
        expanded_lines.append(line)
```

```python
xs = []
cycles = []
x = 1

for cycle in range(len(expanded_lines)):
    
    cycles.append(cycle+1)
    instruct = expanded_lines[cycle]
    
    if instruct == 'noop':
        pass
    
    elif instruct[0:4] == 'addx':
        x = x + int(instruct[5::])
    
    xs.append(x)
               
```

```python
signal_strengths = [x * cycle for x, cycle in zip(xs[0:-1], cycles[1::])]      
```

```python
df = pd.DataFrame({'cycle': cycles[1::], 'x': xs[0:-1], 'signal strength': signal_strengths})
df.set_index('cycle', inplace=True)
```

```python
sum(df.loc[[20, 60, 100, 140, 180, 220]]['signal strength'])
```

```python
def get_signal_cycle(cycles, df):
    
    sum_signals = 0
    
    for cycle in cycles:
        
        sum_signals += df.loc[cycle]['signal strength']
        
    return sum_signals
```

```python
def get_sprite_position(cycle):
    
    if cycle == 1:
        sprite_position = '###.....................................'
    
    else:
        x = df.loc[cycle].x
        sprite_position_lst = list('........................................')
        sprite_position_lst[x-1:x+1] = '###'
        sprite_position = ''.join(sprite_position_lst)
    
    return sprite_position
```

```python
sprite_position = '###.....................................'
pixel_position = 0

for i_start in range(0,len(df),40):
    
    CRT_row = ''

    for cycle in range(i_start+1, i_start+41):
        sprite_position = get_sprite_position(cycle)
        CRT_row += sprite_position[pixel_position-i_start]
        pixel_position+=1

    print(CRT_row)
```
